## 1.1 Difference between Simple Linear Regression and Multiple Linear Regression

**Simple Linear Regression:** This analyzes the relationship between two variables: one predictor (independent) and one outcome (dependent).

**Multiple Linear Regression:** This analyzes the relationship between one dependent variable and two or more independent variables that act as a predictor.

## 1.2 Difference between using a continuous variable and an indicator variable in Simple Linear Regression; and these two linear forms

### Continuous Variable
By using continuous variable, we can use continuous numerical values (e.g., 1.5, 2.7) and the Œ≤1 represents the rate of change in y with respect to x.

**Equation:** y = Œ≤0 + Œ≤1x + œµ

### Indicator Variable
By using indicator variable, we can only use binary (0 or 1) to represent categories (e.g., 1 for male, 0 for female) and the Œ≤1 represents the average difference in y between two groups.

### Equation
y = Œ≤0 + Œ≤1x + œµ

## 1.3 Behavior of the model when a single indicator variable is introduced alongside a continuous variable to create a multiple linear regression

When a single indicator variable is introduced, a **different intercept** will occur for different groups due to the variable indicator shifts the intercept of the regression line depending on category. 

When z=0: The model simplifies to ùë¶ = ùõΩ0 + ùõΩ1ùë• + ùúñ<br>
When z=1: The model becomes ùë¶ = (ùõΩ0 + ùõΩ2) + ùõΩ1ùë• + ùúñ

The **rate of change (slope) across categories** will remain constant, representing  the effect of ùë• on ùë¶ in both groups, assuming no interaction term is added.

**Equation:** y = Œ≤0 + Œ≤1x + Œ≤2z + œµ

## 1.4 The effect of adding an interaction between a continuous and an indicator variable in Multiple Linear Regression models

By adding an interaction between a continuous and an indicator variable in multiple linear regression models, it allows the slope of the continuous variable to differ among categories, changing the model's behavior by creating **different linear trends** for esch category defined by the indicator variable.

**Equation:** y = Œ≤0 + Œ≤1x + Œ≤2z + Œ≤3(x‚ãÖz) + œµ

## 1.5 The behavior of a Multiple Linear Regression model based only on indicator variables derived from a non-binary categorical variable; this linear form; and the necessarily resulting binary variable encodings it utilizes

### Encoding
When a categorical variable with ùëò categories is used in a regression model, it must be encoded into binary (indicator) variables. This is commonly done using dummy encoding, which creates a set of binary variables that represent each category of the original categorical variable. However, to avoid multicollinearity, ùëò‚àí1 binary variables are created, with one category serving as a **reference category.**

### Behavior of the Model
**1. Group-Specific Means:** Each category is associated with a unique mean value of ùë¶. The model estimates a separate mean for each category, reflecting the effect of that category on the dependent variable relative to the reference category.<br>
**2. No Linear Trend:** Since all predictors are categorical and represented by indicator variables, the model does not imply any linear or continuous relationship. Instead, it accounts for categorical shifts between different levels of the non-binary factor.<br>
**3. Interpretation Relative to Reference Category:** The intercept Œ≤0 represents the mean of ùë¶ for the reference category. Each coefficient Œ≤i (for ùëñ = 1, ..., ùëò‚àí1) represents the difference in mean ùë¶ between the corresponding category and the reference category.

### Equation
y = Œ≤0 + Œ≤1z1 + Œ≤2z2 + ... + Œ≤(k‚àí1)z(k‚àí1) + œµ

### 3.5a Additive

```python
import numpy as np
import plotly.graph_objects as go
import statsmodels.formula.api as smf
import pandas as pd

# Fit the model
additive_model = smf.logit('depressed ~ transit + healthcare + shops', data=df).fit()

# Select a few representative values of healthcare and shops
healthcare_values = np.linspace(df['healthcare'].min(), df['healthcare'].max(), 3)  # Fewer values, e.g., min, median, max
shops_values = np.linspace(df['shops'].min(), df['shops'].max(), 3)  # Fewer values

# Create a grid of transit values
transit_values = np.linspace(df['transit'].min(), df['transit'].max(), 100)

# Generate predictions and plot them
fig_additive = go.Figure()

# Add data points
fig_additive.add_trace(go.Scatter(x=df['transit'], y=df['depressed'], mode='markers', name='Data Points'))

# Generate and add best fit lines
for h in healthcare_values:
    for s in shops_values:
        predictions = additive_model.predict(pd.DataFrame({
            'transit': transit_values,
            'healthcare': [h] * 100,
            'shops': [s] * 100
        }))
        fig_additive.add_trace(go.Scatter(
            x=transit_values,
            y=predictions,
            mode='lines',
            name=f'Health: {h:.2f}, Shops: {s:.2f}'
        ))

# Update layout
fig_additive.update_layout(
    title="Additive Model: Best Fit Lines",
    xaxis_title="Transit",
    yaxis_title="Depressed",
    legend_title="Healthcare and Shops"
)

fig_additive.show(renderer='png')
```

```python
import numpy as np
import plotly.graph_objects as go
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split

num_iterations = 100
in_sample_r2 = []
out_of_sample_r2 = []

for i in range(num_iterations):
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)

    model7_spec = smf.ols(formula=(
        'HP ~ scale(center(Attack)) * scale(center(Speed))'
        ' * scale(center(Q("Sp. Def"))) * scale(center(Q("Sp. Atk")))'
        ' + I(Q("Type 1")=="Normal")'
        ' + I(Q("Type 1")=="Water")'
        ' + I(Generation==2)'
        ' + I(Generation==5)'
    ), data=pokeaman_train)
    model7_fit = model7_spec.fit()

    in_sample_r2.append(model7_fit.rsquared)

    y_pred = model7_fit.predict(pokeaman_test)
    y_test = pokeaman_test['HP']
    out_of_sample_r2.append(np.corrcoef(y_test, y_pred)[0, 1]**2)

fig = go.Figure()

fig.add_trace(go.Histogram(
    x=in_sample_r2,
    name="In-Sample R-squared",
    opacity=0.7,
    marker=dict(color="blue"),
    xbins=dict(size=0.01),
))

fig.add_trace(go.Histogram(
    x=out_of_sample_r2,
    name="Out-of-Sample R-squared",
    opacity=0.7,
    marker=dict(color="green"),
    xbins=dict(size=0.01),
))

fig.update_layout(
    title="Distribution of In-Sample and Out-of-Sample R-squared Values for model7_linear_form_CS",
    xaxis_title="R-squared",
    yaxis_title="Frequency",
    barmode="overlay",
    legend=dict(x=0.75, y=0.85),
    bargap=0.1,
)

fig.show(renderer='png')
```

#### Result
The **in-sample $R^2$** values tend t ocluster high, meaning that the variance in target data is explained well when fitting the model. However, the range of values suggests some instability in the fit, which can be due to interactions and complex relationships in the model. Meanwhile, the **out-of-sample $R^2$** has a wider spread, indicating that the model performance varies significantly when tested on the unseen data. Lower **out-of-sample $R^2$** values indicate that the model's predictive power on new data is not as reliable as on training data. This circumstance may arises due to inherent randomness in sampling that usually occurs in small or noisy dataset.

#### Purpose of this demonstration
This demonstration shows how the model‚Äôs performance can vary across different splits of the data, especially for complex models like `model7_linear_form_CS` with interaction terms, illustrating that while a model might perform well on the training set, this performance does not always hold for unseen data.
