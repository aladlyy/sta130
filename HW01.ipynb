7.3 It is necessary to use del df['col'] first before df.dropna() because we can avoid removing important data for our analysis. Let's say that we use df.dropna() first, it will remove every rows that has missing value; in the column "deck", many rows will be removed since there are many missing values there, making us to remove important data from other columns as well. However, if we use del df['col'] to remove column "deck" first, then applying df.dropna(), the only rows that will be deleted are those that are not too important, which has missing values in it.

```python
df.isna().sum()
```

```python
del df['deck']
df = df.dropna()
df.isna().sum()
```
